```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating Fake Sensor Data

```python
N_SAMPLES = 500
```

```python
input_range = np.linspace(0,2*np.pi, N_SAMPLES, dtype=np.float_)
```

```python
signal = np.sin(input_range)
```

```python
noise = np.random.normal(loc=0.0, scale=1, size=N_SAMPLES)
```

```python
assert noise.shape == input_range.shape
```

```python
noisy_signal = signal + noise/5
```

```python
plt.plot(input_range, noisy_signal, 'b.', alpha = 0.5)
plt.plot(input_range, signal, 'r-', linewidth=4)
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel('Time')
plt.ylabel('Value of thing we care about')
plt.show()
```

# Filtering Noise

```python
del signal
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess # https://www.statsmodels.org/stable/generated/statsmodels.nonparametric.smoothers_lowess.lowess.html
```

```python
filtered = lowess(noisy_signal, input_range, frac=0.1)
```

```python
plt.plot(input_range, noisy_signal, 'b.', alpha=0.5)
plt.plot(filtered[:, 0], filtered[:, 1], 'r-', linewidth=4)
plt.legend(['Sensor readings', 'Reconstructed signal'])
plt.show()
```
